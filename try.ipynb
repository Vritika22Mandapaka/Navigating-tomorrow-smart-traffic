```python
import cv2
import torch
from torchvision import models, transforms
import cityscapesscripts.helpers.labels as labels  # Import for class labels

# Load the pre-trained model from Cityscapes scripts
model = torch.hub.load('facebookresearch/semi-supervised-ImageNet1K-models', 'resnet50_swsl')  # Replace with your model if different
model.fc = nn.Linear(2048, 2)  # Adjust output layer for 2 classes
model.eval()

# Define image preprocessing (adjust if needed for your model)
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Get class labels from Cityscapes
class_labels = labels.id2label

# Capture images from the laptop camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if frame is None:
        continue

    # Preprocess the frame
    frame_tensor = transform(frame).unsqueeze(0)

    # Make prediction
    with torch.no_grad():
        output = model(frame_tensor)

    _, predicted = torch.max(output.data, 1)

    # Display result on the frame
    label = class_labels[predicted.item()]  # Use Cityscapes class label
    cv2.putText(frame, f'Prediction: {label}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    cv2.imshow("Image", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release
cv2.destroyAllWindows()
```

# Function to handle client connections
def handle_client(client_socket):
    try:
        while True:
            # Receive command from Colab code
            command = client_socket.recv(1).decode()

            if not command:
                # Break out of the loop if the command is empty (socket closed)
                break

            # Send command to Arduino
            arduino.write(command.encode())

            # Control Arduino lights based on command
            if command == 'E':
                # Turn on green light for emergency
                # Turn off red light
                print("Emergency Vehicle Detected: Turning on green light")
            elif command == 'N':
                # Turn on red light for non-emergency
                # Turn off green light
                print("Non-Emergency Vehicle Detected: Turning on red light")

            # Log that a request has been sent to Arduino
            print(f"Sent command to Arduino: {command}")

    except Exception as e:
        print(f"Error in handle_client: {e}")

    finally:
        # Close the client socket
        client_socket.close()

# Set up the local server
local_server_address = ('127.0.0.1', 9001)
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(local_server_address)
server_socket.listen(1)  # Allow only one connection at a time

print(f"Local server listening on {local_server_address}")

try:
    while True:
        # Wait for a connection from the Colab code
        client_socket, client_address = server_socket.accept()
        print(f"Accepted connection from {client_address}")

        # Start a new thread to handle the client
        client_thread = threading.Thread(target=handle_client, args=(client_socket,))
        client_thread.start()

except KeyboardInterrupt:
    print("Local server terminated by user.")
finally:
    # Close sockets and serial connection
    server_socket.close()
    arduino.close()

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from torchvision.datasets import ImageFolder
from torch.autograd import Variable
import cv2
from google.colab.patches import cv2_imshow


# Define a custom dataset class
class CustomDataset(Dataset):
    def __init__(self, data_dir, transform=None):
        self.data = ImageFolder(root=data_dir, transform=transform)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

data_dir = '/content/train'
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

dataset = CustomDataset(data_dir, transform=transform)
# Split the dataset into training and validation sets
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# Define data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Load pre-trained ResNet18 model
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(512, 2)  # Modify the last fully connected layer for 2 classes (emergency and non-emergency)
model = model.cuda()  # Move model to GPU if available

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    for inputs, labels in train_loader:
        inputs, labels = Variable(inputs.cuda()), Variable(labels.cuda())
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Validation loop
    model.eval()
    correct = 0
    total = 0
    for inputs, labels in val_loader:
        inputs, labels = Variable(inputs.cuda()), Variable(labels.cuda())
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    accuracy = correct / total
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Validation Accuracy: {accuracy:.2%}')

# Test the model on real-time video from the laptop camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    # Preprocess the frame
    frame = cv2.resize(frame, (224, 224))
    frame = transforms.ToTensor()(frame)
    frame = frame.unsqueeze(0).cuda()

    # Make prediction
    model.eval()
    with torch.no_grad():
        output = model(frame)

    _, predicted = torch.max(output.data, 1)

    # Display the result on the frame
    label = "Emergency Vehicle" if predicted.item() == 1 else "Non-Emergency Vehicle"
    cv2.putText(frame, f'Prediction: {label}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Display the resulting frame
    cv2_imshow(frame)

```
